```python
#이제 심화과정을 할거임.
#numpy에서 생성부분을 안했는데, 왜냐면, 우리는 이미 만들어진 데이터를 불러오는게 중요하다보니까.
#load
#파이썬에서는 이미 만들어진 file 을 불러오는게 open 하나뿐이다.
#open 이 불러올수있는 데이터 타입은 2가지 뿐이다. text / binary 이다. 
# 즉, 객체의 정보를 불러올수는 없다는거다. 
# 이게 무슨 뜻이냐면
x = [1,2,3,4,5] #객체지향의 리스트다.
```

```python
with open('list.p','w') as f:
     f.write(x)
#wb는 write binary . 저장한다는 뜻이다. 
#str 이 아니라서 저장할 수 없다고 한다 => 기본적으로 문자열만 저장 가능하다는거다. 
# 문자열 종류는 str / bytes 가 있다
```

```python
x = b'abc'
```

```python
with open('list.p','wb') as f:
     f.write(x)
```

```python
#한마디로 불러들이고 읽는거는 문자열 2종류만 가능한거다. str / bytes
#그냥 저장할수가 없는데,
#pickle 을 쓰면 된다고 한다. 
```

```python
x = [1,2,3,4]
```

```python
import pickle #serializable 직렬화 라고 한다. 객체를 저장해서 객체를 불러오는게 직렬화라고 한다.
```

```python
with open('list.p','wb') as f:
     pickle.dump(x, f)
#x라는 애를 f라는 거에 dump 해라는 뜻이다
```

```python
with open('list.p','rb') as f:
     xx = pickle.load(f)
    #f로 불러오고 xx 로 load 한다는뜻이다
```
