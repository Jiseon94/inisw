```python
N = 100
X = (np.random.rand(N)+np.random.rand(N))*np.pi
Y = np.zeros(N)

Y[np.sin(X) > 0] = 1
```

```python
#1차 미분한 식을 만들어주자.
dLogistic = lambda x, y, theta: x.T.dot(y-logistic(x.dot(theta)))
```

```python
_X = np.c_[np.ones(N), X]
```

```python
h = 1e-5
epoch = 300000

theta = np.random.rand(2)
#x0 = 1, x => vector (x제로는 항상 1이 될거고 x 는 vector 가 될거다.)

for i in range(epoch):
    v = dLogistic(_X, Y, theta)
    u = v/np.linalg.norm(v)
    theta = theta + h*u
```

```python
J = lambda x,t,theta:t*np.log(logistic(x.dot(theta)))+\
                 (1-t)*np.log(1-logistic(x.dot(theta)))
```

```python
h = 1e-5
epoch = 1000000

theta = np.random.rand(2)
history = list()

for i in range(epoch):
    v = dLogistic(_X, Y, theta)
    u = v/np.linalg.norm(v)
    theta = theta + h*u
    
    if i % 1000 ==0:
        history.append(J(_X,Y,theta))
```
