```python
from sklearn.model_selection import cross_val_score
#점수 매기는거. = 기준이 있다는거(성능 측정하는거에)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#R2스퀘어
cross_val_score(KNeighborsClassifier(), iris.iloc[:,:-1], iris.species, cv=10)
#y값 안넣어도 되는 이유 : unsupervised learning 에 쓸 수 있으니 . 
#사이킷런은 넘파이로 만들어졌고, 넘파이는 동시 실행 가능하다. = cpu를 동시실행가능하다.
#n-jobs 는 가동할 cpu 개수를 정할수 있는거다. -1 하면, 내가 가진 최대한을 다 써라는 뜻이다. 

#10등분해서, 테스트한다는거임. 
#근데 편차가 있다. 데이터의 특성이 작으니까 이런 편차가 나는거다. 
#해석 : 86퍼센트 이상은 성능이 나올것 같다는 추측이 가능하다.
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cross_val_score(LogisticRegression(solver='newton-cg'), iris.iloc[:,:-1], iris.species, cv=10)
```

```python
from sklearn.model_selection import StratifiedKFold
#비율에 맞게 쪼개는거.
```

```python
stf = StratifiedKFold(10, shuffle=True)
#stf 를 숫자 대신 슬 수 있다. 
```

```python
cross_val_score(KNeighborsClassifier(), iris.iloc[:,:-1], iris.species, cv=stf)
#cv에 숫자 대신 객체를 집어넣을 수 있다. <- 덕 타이핑 원리로... 이런게 가능하다.데이터 타입 안보고, 이런 기법이 가능한거다. 
```

```python
#abstract method 가 정해져있기에, 우리만의 방식으로 쪼개는거 가능하다. 
```

### 하이퍼 파라미터

```python
temp = []
for i in range(3, 23,2):
    t = cross_val_score(KNeighborsClassifier(i), iris.iloc[:,:-1], 
                       iris.species, cv=5)
    temp.append(t.mean())
```

```python
temp
# k=7일때 제일 성능이 좋다 여겨야 한다. k=9일때도 같은데, 오캄의 면도날이라서 가장 간단한걸 쓴ㄴ거임.
```

```python
# KNeighborsClassifier
temp = []
for i in range(3, 23,2):
    for j in ['uniform', 'distance']:
        t = cross_val_score(KNeighborsClassifier(i, weights=j), iris.iloc[:,:-1], 
                           iris.species, cv=5)
    temp.append([i,j,t.mean()])
```

```python
temp
#알고리즘 이해못해도 이렇게 옵션값 넣어서 다 알 수 있는거다... 알고리즘 몰라도 될까?
```

### Grid search

```python
from sklearn.model_selection import GridSearchCV
```

```python
k = KNeighborsClassifier()
```

```python
k.get_params()
#하이퍼 파라미터 이름을 알 수 있다. 
```

```python
grid = GridSearchCV(KNeighborsClassifier(), { 'n_neighbors': range(3,23,2),
                                             'weights': ['uniform', 'distance'] }, 
                    cv=3, 
                    return_train_score=False)
```

```python
grid.fit(iris.iloc[:,:-1], iris.species)
```

```python
grid.cv_results_
```

```python
#판다스로 집어넣어서 보자
pd.DataFrame(grid.cv_results_).T
#정확도 기준으로 성능을 랭킹까지 매겨서 보여줌.
```

```python
#근데 더 간단하게 보고 싶다면?
grid.best_estimator_
#여러 알고리즘 비교 가능하다.
```

```python
grid.best_index_
```

```python
grid.best_params_
```

```python
grid.best_score_
```
