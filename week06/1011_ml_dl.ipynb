```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
```

```python
knn3 = KNeighborsClassifier(3)
knn5 = KNeighborsClassifier(5)
#하나의 종류에도 이렇게 다양한 이웃 수를 정해서 종류가 다양하게 있을 수있다.
svm = SVC()
#알고리즘을 안다는건, 하이퍼파라미터를 알고 공부한다는것이다.
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(breast.iloc[:,:-1], breast.target)
```

```python
vars(svm)
#svm 은 학습속도가 굉장히 느린아이다. 벡터라이제이션 연산 안하기에. GPU 로 해야함. 
```

```python
svm.fit(X_train, y_train)
#fit 하는 순간 evaluation - optimization 한다.
```

```python
vars(svm)
#학습 한 이후의 데이터다. eval 하고, opti 한 결과이다. 
#성능의 영향력이 다 같진 않다. 팁은, 상위에 있는 항목이 가장 큰 영향력을 가진 편이다.
```

```python
#evaluation 하는 데에도 목표를 달리 한다. 근데, 이건 나중에 배울거다. 
```

```python
svm.score(X_test, y_test)
#이걸로 오버피팅 알 수 없다
```

```python
from sklearn.model_selection import learning_curve
#중요한건 다 model_selection 에 있다.
```

```python
learning_curve(SVC(), X_train, y_train)
```

```python
len(learning_curve(SVC(), X_train, y_train))
```

```python
train_size, train_acc, test_acc = learning_curve(SVC(), X_train, y_train)
```

```python
import sklearn_evaluation
#쉽고 예쁘게 그래프 그리려고
```

```python
sklearn_evaluation.plot.learning_curve(train_acc, test_acc, train_size)
#사이킷런의 결과를 그래프(plot)로 쉽고 예쁘게 그릴 수 있게 해준다.
# 단 아쉬운건 매개변수에 순서가 다르게 들어가야한다.

#왜 나... 그래프가 또 안보이지....
```
